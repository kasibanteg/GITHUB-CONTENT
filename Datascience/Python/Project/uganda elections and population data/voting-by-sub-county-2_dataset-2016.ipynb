```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#hide warnings
import warnings
warnings.filterwarnings('ignore')
#load dataset
df=pd.read_csv("files/voting-by-sub-county-2.csv")
```

```python
df_elections_results=df[['abed_bwanika','amama_mbabazi','baryamureeba_venansius','benon_biraaro','kizza_besigye','joseph_mabirizi','moureen_kyalya','yoweri_museveni','invalid','registered_voters']].sum()
#pd.DataFrame(sum_up_table)
df_elections_results_final=pd.DataFrame([df_elections_results],columns=['abed_bwanika','amama_mbabazi','baryamureeba_venansius','benon_biraaro','kizza_besigye','joseph_mabirizi','moureen_kyalya','yoweri_museveni','invalid','registered_voters'])

#df_elections_results_final

#hide index
#df_elections_results_final2=df_elections_results_final.style.hide_index()

#set styles
# dictionary of column colors
coldict = {'registered_voters':'#ff9a9e', 'invalid':'#fad0c4'}

def highlight_cols(s, coldict):
    if s.name in coldict.keys():
        return ['background-color: {}'.format(coldict[s.name])] * len(s)
    return [''] * len(s)
df_elections_results_final.style.apply(highlight_cols, coldict=coldict)
```

```python
#Total Participats in overall voting process
#df_elections_results_final.iloc[:,0:9]
total_participants_in_voting_execise=df_elections_results_final.iloc[:,0:9].sum(axis=1)
#total_participants_in_voting_execise
#format value and print
print('{:,.0f}'.format(int(total_participants_in_voting_execise)) +' : Is the Total Participats in overall voting process')

```

```python
#people who didnot vote due to reasons are:

#extract registererd voters from the dataset
registered_voters=df_elections_results_final.iloc[:,-1]
#registered_voters
df_not_vote=registered_voters-total_participants_in_voting_execise
df_not_vote

print('{:,.0f}'.format(int(df_not_vote)) +' : are people who were registered but didnot vote')
```

```python
#def get_election_results():
get_ilocs_abed=int(df_elections_results_final.iloc[:,0])
get_ilocs_amama=int(df_elections_results_final.iloc[:,1])
get_ilocs_baryamureeba=int(df_elections_results_final.iloc[:,2])
get_ilocs_benon=int(df_elections_results_final.iloc[:,3])
get_ilocs_besigye=int(df_elections_results_final.iloc[:,4])
get_ilocs_mabirizi=int(df_elections_results_final.iloc[:,5])
get_ilocs_kyalya=int(df_elections_results_final.iloc[:,6])
get_ilocs_museveni=int(df_elections_results_final.iloc[:,7])
get_ilocs_df_not_vote=int(df_not_vote)
    
#pie chart settings    
pie_graph_data=[get_ilocs_abed, get_ilocs_amama, get_ilocs_baryamureeba,get_ilocs_benon,get_ilocs_besigye,get_ilocs_mabirizi,get_ilocs_kyalya,get_ilocs_museveni]
pie_graph_labels=["Abed Bwanika", "Amama Mbabazi", "Baryamureeba Venansius", "Benon Biraaro","Kizza Besigye","Joseph Mabirizi","Moureen Kyalya","Yoweri Museveni"]
pie_graph_colors=['green','orange','brown','lightgreen','blue','grey','darkgrey','yellow']
#bar graph settings
bar_graph_data=[get_ilocs_abed, get_ilocs_amama, get_ilocs_baryamureeba,get_ilocs_benon,get_ilocs_besigye,get_ilocs_mabirizi,get_ilocs_kyalya,get_ilocs_museveni,get_ilocs_df_not_vote]
bar_graph_labels=["Abed Bwanika", "Amama Mbabazi", "Baryamureeba Venansius", "Benon Biraaro","Kizza Besigye","Joseph Mabirizi","Moureen Kyalya","Yoweri Museveni","Not Voted"]
bar_graph_colors=['green','orange','brown','lightgreen','blue','grey','darkgrey','yellow','darkred']
```

```python

#import matplotlib.pyplot as plt
sizes = pie_graph_data
labels = pie_graph_labels

fig1, ax1 = plt.subplots(figsize=(6, 5))
fig1.subplots_adjust(0.3,0,1,1)

#theme = plt.get_cmap('bwr')
#ax1.set_prop_cycle("color", [theme(1. * i / len(sizes)) for i in range(len(sizes))])

color = pie_graph_colors
ax1.set_prop_cycle(color=color)

_, _ = ax1.pie(sizes, startangle=90)

ax1.axis('equal')

total = sum(sizes)
plt.legend(
    loc='upper right',
    labels=['%s, %1.1f%%' % (
        l, (float(s) / total) * 100) for l, s in zip(labels, sizes)],
    prop={'size': 11},
    bbox_to_anchor=(0.0, 1),
    bbox_transform=fig1.transFigure
)

plt.show()
```

```python
#plot bar graph
#https://bmanohar16.github.io/blog/customizing-bar-plots
plt.barh(bar_graph_labels,bar_graph_data,color=bar_graph_colors)
plt.title("dddd")
plt.show()
```

```python
# Import libraries
import pandas as pd

from matplotlib import pyplot as plt
%matplotlib inline
from matplotlib import rcParams

# Custom Font
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Verdana']
rcParams['font.size'] = 14

# Figure Size
fig, ax = plt.subplots(figsize=(10,7))

# Horizontal Bar Plot
ax.barh(bar_graph_labels,bar_graph_data , color=bar_graph_colors)

# Remove axes splines
for s in ['top','bottom','left','right']:
    ax.spines[s].set_visible(False)

# Remove x,y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad=5)
ax.yaxis.set_tick_params(pad=10)

# Add x,y gridlines
ax.grid(b=True, color='grey', linestyle='-.', linewidth=0.5, alpha=0.2)

# Show top values 
ax.invert_yaxis()

# Add Plot Title
ax.set_title('Uganda Presidential Results : 2016',
             loc='left', pad=10)

# Add annotation to bars

for i in ax.patches:
    ax.text(i.get_width()+500, i.get_y()+0.5, str(round((i.get_width()), 2)),
            fontsize=10, fontweight='bold', color='grey')
    

# Add Text watermark
fig.text(0.9, 0.15, 'www.kasibante.com', fontsize=12, color='grey',
         ha='right', va='bottom', alpha=0.5)

# Save Plot as image
fig.savefig('Uganda Presidential Results.png', dpi=100,
            bbox_inches='tight')

# Show Plot
plt.show()
```

## Leader per district

```python
#load summarised data set by Leader per district...
#final-vote-by-district-2016-presidential.csv
```

```python
leaderPerDistrictDataSet=pd.read_csv("files/final-vote-by-district-2016-presidential.csv")
leaderPerDistrictDataSet
```

```python
#new_column:: split only District names from DISTRICT COLUMN
leaderPerDistrictDataSet['new_districts']=(leaderPerDistrictDataSet.DISTRICT).str.split(" ",expand=True)[1]
leaderPerDistrictDataSet
```

```python
#load all world cities cordinates
world_cities_cordinates=pd.read_csv("files/worldcities.csv")
world_cities_cordinates
```

```python
##filter only Ugandan cities
uganda_filtered_cities=world_cities_cordinates.loc[(world_cities_cordinates.country == 'Uganda'), ['city_ascii','lat','lng','country']]
#convert column city_ascii to upper case to match with the leaderPerDistrictDataSet
uganda_filtered_cities['city_ascii'] = uganda_filtered_cities['city_ascii'].str.upper()
uganda_filtered_cities
```

```python
##Join leaderPerDistrictDataSet and uganda_filtered_cities---two Data sets
joined_table=pd.merge(leaderPerDistrictDataSet,uganda_filtered_cities, how='left', left_on='new_districts',right_on='city_ascii')
joined_table
```

```python
joined_table.info()
```

```python
#we need to fill in the nan values for the joined_table dataset

# 1. for city_ascii column
joined_table.loc[joined_table.city_ascii.isnull(), 'city_ascii']=joined_table.new_districts
joined_table


#Get all rows with missing lat,lng,country
# there 15 rows so we need to fill in the missing data manually
nan_values = joined_table[joined_table['lat'].isna()]
nan_values


```

```python
#2. latitude coordinates
#{KABAROLE,}
#options=['0.589668']
#filled=nan_values.loc[nan_values.city_ascii.isnull(), 'lat']=options
####https://www.distancesto.com/coordinates/ug/kabarole-latitude-longitude/history/67234.html
```

```python
nan_values.loc[nan_values.new_districts == "KABAROLE", ["lat","lng"]] = ["0.589668","30.254879"]
nan_values.loc[nan_values.new_districts == "KIBAALE", ["lat","lng"]] = ["0.768255","31.077659"]
nan_values.loc[nan_values.new_districts == "LUWEERO", ["lat","lng"]] = ["0.840409","32.497668"]
nan_values.loc[nan_values.new_districts == "SSEMBABULE", ["lat","lng"]] = ["-0.081638","31.459901"]
nan_values.loc[nan_values.new_districts == "NAKASEKE", ["lat","lng"]] = ["1.223085","32.083744"]
nan_values.loc[nan_values.new_districts == "BUDAKA", ["lat","lng"]] = ["1.015512","33.949915"]
nan_values.loc[nan_values.new_districts == "BULIISA", ["lat","lng"]] = ["2.112727","31.417046"]
nan_values.loc[nan_values.new_districts == "BUVUMA", ["lat","lng"]] = ["0.092168","33.200464"]
nan_values.loc[nan_values.new_districts == "GOMBA", ["lat","lng"]] = ["0.159116","31.673937"]
nan_values.loc[nan_values.new_districts == "LUUKA", ["lat","lng"]] = ["0.72506","33.303714"]
nan_values.loc[nan_values.new_districts == "BUTAMBALA", ["lat","lng"]] = ["0.17425","32.106467"]
nan_values.loc[nan_values.new_districts == "SHEEMA", ["lat","lng"]] = ["-0.55153","30.389665"]
nan_values.loc[nan_values.new_districts == "KWEEN", ["lat","lng"]] = ["1.412581","34.552839"]
nan_values.loc[nan_values.new_districts == "MITOOMA", ["lat","lng"]] = ["-0.619328","30.020296"]
nan_values.loc[nan_values.new_districts == "BUHWEJU", ["lat","lng"]] = ["-0.356177","30.29742"]
nan_values.country.fillna('Uganda',inplace=True)
nan_values
```

```python
#dropping nan values from joined_table and join with nan_values row--wise
# drop nan values
joined_table.dropna(inplace=True)
#joining two tables
final_data_set=pd.concat([joined_table,nan_values],axis=0)
final_data_set
```
