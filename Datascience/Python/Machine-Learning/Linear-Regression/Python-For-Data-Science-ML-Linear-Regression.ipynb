 - Identifying Independent and Dependent variables
 
 
 THUMB RULE FOR Machine Learning Models
 1) INCLUDE ONLY INDEPENDENT VARIABLES THAT ARE HIGHLY CORRELATED WITH DEPENDENT VARIABLES
   1.1) IF THRE ARE ANY INDEPENT VARIABLES THAT ARE HIGHLY CORRELATED WITH EACH OTHER THEN
         1.1.1) INCLUDE ONLY ONE OF THOSE INDEPENDENT VARIABLES IN YOUR MODEL 
 
 2) ALL INDEPENDENT VARIABLES SHOULD NOT BE CORRELATED WITH EACH OTHER. THEY SHOULD BE CORRELATED ONLY WITH THE DEPENDENT
  CLEAR
         
         For example if
          nox,crim,ptration,lstat,b are all high correelated with price
          
          1) Include all of them firm
            1.1) remove crim, and include nox,ptratio,lstat,b  with price  
            or 
            1.2) remove nox and include crim,ptratio,lstat,b  with price 
 
 Correlation#1 - Relevancy check = find relationship between ind and dep variable
 Find out those Independent variables that are highly correlated (both +ve and -ve) with Dependent variable
 - what are independent variables that will be highly impacting "medv" 
 - these ind variables will be included in model building process where we split the data into training and test data
 
 but the below is also important
 Correlation#2 - Multicollinearity - find the relationship between 2 independeent variables
 Find out those Independent variables that are highly correlated (both +ve and -ve) with another 
                       InDependent variable
                       
 - avoid overfitting or underfitting
 
 
 Relevancy check helps in identifying those variables (independent and dependent) for the model building process
 
 Multicollinearity helps in identify those independent variables which are highly correlated and 
         - include ONLY one of those indepenent variables in the model building process
         
         if nox and crime are highly correlated - it means they are multicollinear and including them in the same model will result in model bias 
         
          so we do the below
             1) create model with independent variables which are highly correlated with price and ensure that in the list of
                ind variables, have either nox or crime
             2) model1 = nox, lstat,b,ptratio,price
                model2 = crim, lstant,b,ptratio,price
         
         for example assume 
              chas and zn - high correlation
              zn and dis  - low correlation (not related)
              
          model building
              include ivs as chas, dis and dv as medv - model 1
              include ivs as zn,chas and  dv as med   - model 2

# Regression Evaluation Metrics


Here are three common evaluation metrics for regression problems:

**Mean Absolute Error** (MAE) is the mean of the absolute value of the errors:

    1) Take the difference between actual y value and predicted y value for all the rows(observations -500 odd observations)
    2) Add all the differences 
        3) MODULUS will remove the sign
    4) divide by no of observations

$$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$

**Mean Squared Error** (MSE) is the mean of the squared errors:

    1) Take the difference between actual y value and predicted y value for all the rows(observations -500 odd observations)
    2) square the differences
    3) Add all the squared differences 
    4) divide by no of observations

$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

**Root Mean Squared Error** (RMSE) is the square root of the mean of the squared errors:

1) Take the difference between actual y value and predicted y value for all the rows(observations -500 odd observations)
2) square the differences
3) Add all the squared differences 
4) divide by no of observations
5) take a square root

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

Comparing these metrics:


- **MAE** is the easiest to understand, because it's the average error. 
          Take your error, take the absolute value and get an average
          
- **MSE** is more popular than MAE, because MSE "punishes" larger errors, which tends to be useful in the real world.
          More useful in real world. takes into account larger errors as its squaring the errors 

- **RMSE** is even more popular than MSE, because RMSE is interpretable in the "y" units.
           

All of these are **loss functions**, because we want to minimize them. to get the best model , all these errors have to be less
