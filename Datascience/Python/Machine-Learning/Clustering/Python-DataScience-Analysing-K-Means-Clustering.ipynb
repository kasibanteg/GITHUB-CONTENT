## K Means Cluster Creation

Now it is time to create the Cluster labels!

** Import KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

** Create an instance of a K Means model with 2 clusters.**

```python
kmeans = KMeans(n_clusters=2)
```

**Fit the model to all the data except for the Private label.**

```python
kmeans.fit(df.drop('Private',axis=1))
```

** What are the cluster center vectors?**

```python
kmeans.cluster_centers_ 
```

```python
kmeans.labels_
```

```python
kmeans.score
```

```python
kmeans.cluster_centers_.shape
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters, keep in mind, you usually won't have this luxury in the real world.

** Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```python
### applies the function "converter" on every row of the column "Private"
### whenever the value of "Private" = 'Yes', the function converter will convert it to 1 and apply the value to the column
### "Cluster" . Note that the column "Cluster" is newly added to the dataframe
df['Cluster'] = df['Private'].apply(converter)
```

** Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.**

```python
from sklearn.metrics import confusion_matrix,classification_report

#### kmeans_labels is an attribute of the kmeans clustering algorithm. gives the cluster defintion of every 
### observation in the data frame df

### df['Cluster'] is the actual cluster class, prior to running the clustering algorithm
### confusion matrix will give out the accuracy of clustering, by comparing between 
###  df['Cluster'](orginal data) vs kmeans_labels (post running kmeans clustering)
### 
print(confusion_matrix(df['Cluster'],kmeans.labels_))
##print(classification_report(df['Cluster'],kmeans.labels_))
```

```python
## Confusion matrix between original y (y_test) and predicted y (pred)
###### to check the accuracy
print(metrics.accuracy_score(df['Cluster'],kmeans.labels_))  
```
